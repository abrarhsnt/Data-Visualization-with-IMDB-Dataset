**Classify Movies Based on Ratings [Excellent, Good, and Average]**

```python
df.columns
```

```python
#creating our own function 'rating' to classify between Excellent, Good, and Average 

def rating(rating):
  if rating>=7.0:
    return "Excellent"
  elif rating>=6.0:
    return "Good"
  else:
    return "Average" 
```

```python
#applying the created function on our dataset and create a new coloumn named 'Rating Classification'

df['Rating Classification']=df['Rating'].apply(rating)
```

**Find Unique Values From Genre**

```python
df.columns
```

```python
df['Genre']
```

```python
#creating a list with all the genres splitted by ','

list1=[]
for value in df['Genre']:
  list1.append(value.split(','))
```

```python
list1
```

```python
#Converting the 2D list into 1D list

one_d=[]
for item in list1:
  for item1 in item:
    one_d.append(item1)
```

```python
one_d
```

```python
unique_list=[]
for item in one_d:
  if item not in unique_list:
    unique_list.append(item)
```

```python
unique_list
```

```python
len(unique_list)
```
